```python
import matplotlib.pyplot as plt
import numpy as np
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))

# Grfico 1
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))
plt.title("Seno")

# Grfico 2
plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.title("Coseno")

# Ms espacio entre los subplots.
plt.subplots_adjust(wspace=0.5)
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize = (5,4)) 
ax1.plot(x, np.sin(x), color = "red")
ax2.plot(x, np.cos(x), color = "green")
ax3.plot(x, np.tan(x), color = "blue")

for ax in [ax1, ax2, ax3]:
    ax.tick_params(axis='both', labelsize=6)
plt.subplots_adjust(hspace=0.5)
```


### Ejercicio 3: Ajuste de Colores y Estilos de Lnea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
x = np.linspace(0,10,100)
plt.figure(figsize=(10,4))

# Grfico 1
plt.subplot(1,2,1)
plt.plot(x, np.sin(x*2), "--", color = "purple")
plt.title("Seno")

# Grfico 2
plt.subplot(1,2,2)
plt.plot(x, np.cos(x*2), "-.", color = "cyan")
plt.title("Coseno")

# Ms espacio entre los subplots.
plt.subplots_adjust(wspace=0.5)
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python
x = np.linspace(0,10,100)
plt.figure(figsize=(5,4))

# Seno hiperblico.
plt.subplot(2,1,1)
plt.plot(x, np.sinh(x), color = "red")
plt.title("Seno hiperblico", fontsize = 7)
plt.ylabel("sinh(x)", fontsize = 6)
plt.xlim(0,10)
plt.tick_params(labelsize = 6)

# Coseno hiperblico.
plt.subplot(2,1,2)
plt.plot(x, np.cosh(x), color = "blue")
plt.title("Coseno hiperblico", fontsize = 7)
plt.ylabel("cosh(x)", fontsize = 6)
plt.xlim(0,10)
plt.tick_params(labelsize = 6)

# Ajuste entre grficas.
plt.subplots_adjust(hspace=0.5)
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[1,0].plot(x, np.tan(x))
ax[1,1].plot(x, np.log(x))

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

## Estilo "axes"

```python
fig, ax = plt.subplots(2)

# Raz cuadrada y coseno
ax[0].plot(x, np.cos(x), label = "Coseno")
ax[0].set_title("Raz cuadrada y coseno")
ax[0].plot(x, np.sqrt(x), label = "Raz cuadrada")
ax[0].legend()
# Seno
ax[1].plot(x, np.sin(x))
ax[1].set_title("Seno")

plt.subplots_adjust(hspace=0.5)
```

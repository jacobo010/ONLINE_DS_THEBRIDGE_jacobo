```python
import matplotlib.pyplot as plt
import numpy as np
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))

# Gráfico 1
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))
plt.title("Seno")

# Gráfico 2
plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.title("Coseno")

# Más espacio entre los subplots.
plt.subplots_adjust(wspace=0.5)
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize = (5,4)) 
ax1.plot(x, np.sin(x), color = "red")
ax2.plot(x, np.cos(x), color = "green")
ax3.plot(x, np.tan(x), color = "blue")

for ax in [ax1, ax2, ax3]:
    ax.tick_params(axis='both', labelsize=6)
plt.subplots_adjust(hspace=0.5)
```


### Ejercicio 3: Ajuste de Colores y Estilos de Línea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
x = np.linspace(0,10,100)
plt.figure(figsize=(10,4))

# Gráfico 1
plt.subplot(1,2,1)
plt.plot(x, np.sin(x*2), "--", color = "purple")
plt.title("Seno")

# Gráfico 2
plt.subplot(1,2,2)
plt.plot(x, np.cos(x*2), "-.", color = "cyan")
plt.title("Coseno")

# Más espacio entre los subplots.
plt.subplots_adjust(wspace=0.5)
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperbólico y coseno hiperbólico se obtienen así:
* np.sinh, seno hiperbólico;
* np.cosh, coseno hiperbólico;


```python
x = np.linspace(0,10,100)
plt.figure(figsize=(5,4))

# Seno hiperbólico.
plt.subplot(2,1,1)
plt.plot(x, np.sinh(x), color = "red")
plt.title("Seno hiperbólico", fontsize = 7)
plt.ylabel("sinh(x)", fontsize = 6)
plt.xlim(0,10)
plt.tick_params(labelsize = 6)

# Coseno hiperbólico.
plt.subplot(2,1,2)
plt.plot(x, np.cosh(x), color = "blue")
plt.title("Coseno hiperbólico", fontsize = 7)
plt.ylabel("cosh(x)", fontsize = 6)
plt.xlim(0,10)
plt.tick_params(labelsize = 6)

# Ajuste entre gráficas.
plt.subplots_adjust(hspace=0.5)
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[1,0].plot(x, np.tan(x))
ax[1,1].plot(x, np.log(x))

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

Reproduce la figura siguiente empleando sólo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros gráficos (fila 1). Luego usa el estilo con métodos (objetos `axes`) para añadir el tercer gráfico.

## Estilo "matlab"

```python
plt.figure(figsize=(7,7))

# Gráfico coseno
plt.subplot(2,1,1)
plt.plot(x, np.cos(x))
plt.title("coseno")

# Gráfico seno
plt.subplot(2,1,2)
plt.plot(x, np.sin(x))
plt.title("Seno")

plt.subplots_adjust(hspace= 0.5)
```

## Estilo "axes"

```python
fig, ax = plt.subplots(2)

# Raíz cuadrada y coseno
ax[0].plot(x, np.cos(x), label = "Coseno")
ax[0].set_title("Raíz cuadrada y coseno")
ax[0].plot(x, np.sqrt(x), label = "Raíz cuadrada")
ax[0].legend()
# Seno
ax[1].plot(x, np.sin(x))
ax[1].set_title("Seno")

plt.subplots_adjust(hspace=0.5)
```

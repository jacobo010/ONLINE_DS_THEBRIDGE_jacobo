### Ejercicio 9

Usa scipy tal como lo hicimos en el workout para hacer el test de correlación. ¿Cuál es el p-value?¿Qué podemos decir de la correlación lineal entre "total" y "speeding"?

```python
from scipy.stats import pearsonr
pearsonr(df_crashes["total"], df_crashes["speeding"], alternative= "less")
```

Nuestro p-value es prácticamente 1, por lo que nuestras variables pueden tener una correlación lineal y a no rechazamos la hipótesis nula.

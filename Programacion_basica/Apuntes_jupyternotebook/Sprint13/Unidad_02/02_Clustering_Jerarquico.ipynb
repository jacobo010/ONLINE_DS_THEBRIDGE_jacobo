### Visualizando Estructura Jerrquica con Dendrogramas

### Understanding Hierarchical Clustering

Hasta ahora, hemos explorado los datos, codificado las columnas categricas, decidido qu columnas eran adecuadas para el clustering y reducido la dimensionalidad de los datos. Los grficos indican que tenemos 5 clusters en nuestros datos. Hasta ah la forma "no jerarquica" de mirar el problema.   

Pero **tambin hay otra forma de visualizar las relaciones entre nuestros puntos y ayudar a determinar el nmero de clusters: creando un dendrograma** (comnmente mal escrito como dendogram). Dendro significa rbol en latn.  

El dendrograma es el resultado del enlace de puntos en un conjunto de datos. Es una representacin visual del proceso de clustering jerrquico. Y cmo funciona el proceso de clustering jerrquico? Bueno... depende, probablemente una respuesta que ya has escuchado mucho en este bootcamp :-)

### Comprendiendo el Agrupamiento Jerrquico

Cuando el algoritmo de agrupamiento jerrquico (**HCA, Hierarchical Clustering Algorithm**) comienza a vincular los puntos y encontrar grupos, primero puede dividir los puntos en 2 grandes grupos, y luego dividir cada uno de esos dos grupos en grupos ms pequeos, teniendo un total de 4 grupos, lo cual es el enfoque **divisivo** y de **arriba hacia abajo**.

Alternativamente, puede hacer lo contrario: puede mirar todos los puntos de datos, encontrar 2 puntos que estn ms cercanos entre s, vincularlos, y luego encontrar otros puntos que sean los ms cercanos a esos puntos vinculados y seguir construyendo los 2 grupos desde **abajo hacia arriba**. Este es el enfoque **aglomerativo** que desarrollaremos.

### Pasos para realizar el agrupamiento jerrquico aglomerativo

Para hacer an ms claro el enfoque aglomerativo, hay pasos en el algoritmo de Clustering Jerrquico Aglomerativo (AHC):

    1. Al inicio, se trata cada punto de datos como un clster. Por lo tanto, el nmero de clsteres al inicio ser K, donde K es un nmero entero que representa el nmero de puntos de datos.
    2. Se forma un clster uniendo los dos puntos de datos ms cercanos, lo que resulta en K-1 clsteres.
    3. Se forman ms clsteres uniendo los dos clsteres ms cercanos, lo que resulta en K-2 clsteres.
    4. Se repiten los tres pasos anteriores hasta que se forme un nico clster grande.

Si inviertes los pasos del algoritmo ACH, yendo del 4 al 1 - esos seran los pasos para **Divisive Hierarchical Clustering (DHC)**

Otro punto importante es que se introduce el concepto de clsteres cercanos y lejanos, es decir tendremos que fijar una forma de evaluar cuando un clster est ms lejos o ms cerca de otro. Es decir utilizar un concepto de distancia entre clsteres.

### Trazando y entendiendo el Dendrograma

Vamos a trazar el dendrograma de nuestros datos de clientes para visualizar las relaciones jerrquicas de los datos. Esta vez, usaremos la biblioteca scipy para crear el dendrograma de nuestro conjunto de datos:

```python
customer_data.head()
```

```python
customer_data.columns
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.title("Customers Dendrogram")

selected_data = customer_data[["Annual Income (k$)","Spending Score (1-100)"]]
clusters = shc.linkage(selected_data, 
            method='ward', # Este hiperparmetro y el siguiente tienen que ver con la distancia entre clsteres y los veremos un poco ms adelante
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```

En el script anterior, hemos generado los clsteres y subclsteres con nuestros puntos, definido cmo se conectaran nuestros puntos (aplicando el mtodo ward), y cmo medir la distancia entre los puntos (usando la mtrica euclidiana).

Para leer el dendrograma, hay que tener en cuenta que la "y" representa la distancia en los trminos que se le haya dado al mtodo `linkage` (en este caso ward, euclidean) de los diferentes clusters. Si trazamos una lnea horizontal nos dice cuntos clsters hay a una distancia igual o superior. As cortamos por el 0, todos los puntos son clusters. Si cortamos por un valor determinado de distancia se generarn tantos clusters como lnea horizontales queden por debajo o lneas individuales. Si cortamos en la figura de arriba por y (distancia) = 250, quedarn 4 clsters:

Es por es que lo mtodos de vinculacin (definidos por los argumentos que le damos al mtodo linkage) son fundamentales en un clustering jerrquico. Vemoslos... en la siguiente sesin

### Mtodos de vinculacin (Linkage Methods)

Hay muchos mtodos de enlace, al entender ms sobre cmo funcionan, podrs elegir el apropiado para tus necesidades. Adems, cada uno de ellos dar diferentes resultados cuando se apliquen. No hay una regla fija en el anlisis de agrupamiento, si es posible, estudia la naturaleza del problema para ver cul se ajusta mejor, prueba diferentes mtodos e inspecciona los resultados.

Algunos de los mtodos de enlace son:

   '

**Vnculo simple/Simple Linkage:** tambin conocido como vecino ms cercano (NN). La distancia entre los clsteres se define por la distancia entre sus miembros ms cercanos.



**Enlace completo/Complete Linkage:** tambin conocido como Vecino ms lejano (FN), Algoritmo del punto ms lejano, o Algoritmo Voor Hees. La distancia entre los clsteres se define por la distancia entre sus miembros ms lejanos. Este mtodo es computacionalmente costoso.



**Enlace promedio/Average Linkage:** tambin conocido como UPGMA (Mtodo de Agrupamiento de Pares no Ponderado con Media Aritmtica). Se calcula el porcentaje de puntos de cada grupo con respecto al nmero de puntos de los dos grupos si se fusionaran.

   

**Enlace del centroide/centroid linkage:** tambin conocido como UPGMC (Mtodo de Grupo de Pares no Ponderado utilizando Centroides). Se calcula un punto definido por la media de todos los puntos (centroide) para cada clster y la distancia entre clsteres es la distancia entre sus respectivos centroides.



**Enlace de Ward/Ward Linkage:** Tambin conocido como MISSQ (Minimal Increase of Sum-of-Squares). Especifica la distancia entre dos clsteres, calcula el error de la suma de cuadrados (ESS) y elige sucesivamente los siguientes clsteres en funcin del ESS ms pequeo. El Mtodo de Ward busca minimizar el aumento de ESS en cada paso. Por lo tanto, minimizando el error.



En cuanto a la **distancia**, tambin existen varias opciones (Euclidiana, Manhattan, Minkowski, Chebyshev, Coseno, Jaccard, etc).

Hemos elegido Ward y Euclidiana para el dendrograma porque son el mtodo y mtrica ms comnmente usados. Por lo general, dan buenos resultados ya que Ward enlaza puntos minimizando errores, y Euclidiana funciona bien con poca dimensionalidad.

En este ejemplo, estamos trabajando con dos caractersticas (columnas) de los datos de marketing, y 200 observaciones o filas. Dado que el nmero de observaciones es mayor que el nmero de caractersticas (200 > 2), estamos trabajando en un espacio de baja dimensin.

### Seleccin de K a partir del dendrograma

Ya hemos discutido mtricas, enlaces y cmo cada uno de ellos puede impactar en nuestros resultados. Ahora continuemos con el anlisis del dendrograma y veamos cmo puede darnos una indicacin del nmero de clsteres en nuestro conjunto de datos.

Encontrar un nmero interesante de clsteres en un dendrograma es un pequeo arte con varios mtodos posibles de aproximacin. Nostros vamos a aplicar el denominado **mtodo del mayor salto (gap)**. Consiste en buscar el espacio horizontal ms grande (este es el mayor gap) que no tiene lneas verticales (el espacio con las lneas verticales ms largas). 

Podemos trazar una lnea horizontal que pase a travs de esa distancia ms larga:

```python
plt.figure(figsize=(10, 7))
plt.title("Customers Dendogram with line")
clusters = shc.linkage(selected_data, 
            method='ward',
            metric="euclidean")
shc.dendrogram(clusters)
plt.axhline(y = 300, color = 'r', linestyle = '-')
```

Despus de ubicar la lnea horizontal, contamos cuntas veces nuestras lneas verticales fueron cruzadas por ella; en este ejemplo, 3 veces. As que 3 parece una buena indicacin del nmero de agrupaciones que tienen la mayor distancia entre ellas.  

Pero siempre podemos complementar el anlisis con el que vimos del scatter plot de nuestras dos features, y comprobar para 5 clusters que distancia hay entre clusters y si nos parece adecuada entonces nos podemos saltar el mecanismo anterior:


```python
plt.figure(figsize=(10, 7))
plt.title("Customers Dendogram with line")
clusters = shc.linkage(selected_data, 
            method='ward', 
            metric="euclidean")
shc.dendrogram(clusters)
plt.axhline(y = 240, color = 'r', linestyle = '-')
```

Y no parece mala, no tiene tanto gap como 3 clsteres pero se les ve razonablemente separados.

**Nota: El dendrograma solo debera considerarse como una referencia al elegir el nmero de clusters. Puede desviarse fcilmente de ese nmero y est completamente influenciado por el tipo de enlace y mtricas de distancia. Al realizar un anlisis detallado de clusters, se recomienda observar dendrogramas con diferentes enlaces y mtricas, y observar los resultados generados con las primeras tres lneas en las cuales los clusters tienen la mayor distancia entre ellos.**

### Implementando un Clustering Jerrquico Aglomerativo

Hasta ahora hemos calculado el nmero sugerido de clsteres para nuestro conjunto de datos que corroboran con nuestro anlisis inicial. Este paso la eleccin de k o de los hiperparmetros se adelante en este proceso al proceso con datasets supervisados.

Ahora podemos crear nuestro modelo de agrupamiento jerrquico aglomerativo utilizando AgglomerativeClustering de Scikit-Learn y averiguar las etiquetas de los puntos de marketing con labels_:

```python
from sklearn.cluster import AgglomerativeClustering

clustering_model = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
clustering_model.fit(selected_data)
clustering_model.labels_
```

```python
data_labels = clustering_model.labels_
sns.scatterplot(x='Annual Income (k$)', 
                y='Spending Score (1-100)', 
                data=selected_data, 
                hue=data_labels, palette = "rainbow").set_title('Labeled Customer Data')
```

Este es nuestro conjunto de datos final clusterizado. Puedes ver los puntos de datos codificados por colores en forma de cinco grupos.

Los puntos de datos en la esquina inferior derecha (etiqueta: 0, puntos de datos morados) pertenecen a los clientes con salarios altos pero bajos gastos. Estos son los clientes que gastan su dinero cuidadosamente.

De manera similar, los clientes en la esquina superior derecha (etiqueta: 2, puntos de datos verdes) son aquellos con salarios altos y altos gastos. Estos son el tipo de clientes a los que las empresas apuntan.

Los clientes en el medio (etiqueta: 1, puntos de datos azules) son aquellos con ingresos y gastos promedio. El mayor nmero de clientes pertenece a esta categora. Las empresas tambin pueden apuntar a estos clientes dado el hecho de que son numerosos.

Los clientes en la esquina inferior izquierda (etiqueta: 4, rojo) son aquellos que tienen salarios bajos y bajos gastos, podran ser atrados ofrecindoles promociones.

Y finalmente, los clientes en la esquina superior izquierda (etiqueta: 3, puntos de datos naranjas) son aquellos con gastos altos e ingresos bajos, que quiz son los clientes que deberamos evitar.



### Ventajas del agrupamiento jerrquico: Cundo usarlo y cundo no

   - **Obtencin de observaciones ms similares a cualquier observacin dada:** El principal beneficio del agrupamiento jerrquico es que proporciona informacin detallada sobre qu observaciones son ms similares entre s. Este nivel de detalle no es proporcionado por muchos otros algoritmos, que generalmente solo devuelven el ID del clster al que pertenece una observacin dada. El agrupamiento jerrquico es particularmente til en situaciones en las que tienes algunas observaciones en las que ests particularmente interesado y deseas poder identificar observaciones similares a esas observaciones.
   
   - **Poco sensible a condiciones de inicializacin:** Otro beneficio del agrupamiento jerrquico es que no es sensible a condiciones de inicializacin, como semillas establecidas o el orden del conjunto de datos. Generalmente obtendrs resultados muy similares, y en algunos casos el mismo resultado exacto, si vuelves a ejecutar tu anlisis con diferentes condiciones de inicializacin.
   
   - **Puede adaptarse para incorporar variables categricas:** Otro beneficio del agrupamiento jerrquico es que se puede adaptar para admitir situaciones donde tienes una mezcla de variables numricas y categricas relativamente fcilmente. Para hacer esto, debes asegurarte de que ests utilizando una mtrica de distancia apropiada para tipos de datos mixtos, como la distancia de Grower.
    
   - **Bien estudiado:** Despus del agrupamiento k-means, el agrupamiento jerrquico es probablemente el segundo tipo de algoritmo de agrupamiento ms popular y bien estudiado. Eso significa que ms de tus compaeros de trabajo estarn familiarizados con l, lo que har ms fcil para ellos entender y contribuir a los anlisis que utilizan el agrupamiento jerrquico.
   
   - **Menos sensible a valores atpicos:** El agrupamiento jerrquico es menos sensible a valores atpicos que algunos otros algoritmos de agrupamiento. Esto significa que la presencia de unos pocos valores atpicos probablemente no afectar la forma en que el algoritmo funciona en los otros puntos de datos. Esto se debe a que generalmente los valores atpicos no se agregan a un clster hasta el final del proceso, cuando todas las dems observaciones ya han sido manejadas (al menos para el agrupamiento jerrquico aglomerativo).
   
   - **Menos supuestos estrictos sobre la forma de los clsteres:** Los algoritmos de agrupamiento jerrquico no hacen supuestos tan estrictos sobre la forma de tus clsteres. Dependiendo de la mtrica de distancia que uses, algunas formas de clsteres pueden detectarse ms fcilmente que otras, pero hay ms flexibilidad.

### Desventajas del agrupamiento jerrquico

   - **Lento:** Una desventaja del agrupamiento jerrquico es que es relativamente lento. Generalmente requiere el clculo de la distancia entre todas las observaciones en tu conjunto de datos, por lo que el nmero de clculos requeridos crece rpidamente a medida que aumenta el tamao de tu conjunto de datos.
